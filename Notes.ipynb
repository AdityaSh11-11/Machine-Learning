## MACHINE LEARNING 

What is Mchine Learning?

Machine learning is the science of getting computers to learn and act like humans do and improve their learning over time in autonomous manner, by feeding them data and information in the form of observation and real world interactions


Machine learning system learns from historical data,builds the prediction models and whenever it receives new data predicts the output for it.

Historical data {Input Data + Output Data} --> Program(model)

Machine Learning is classified into:-
1. Supervised Learning
2. UnSupervised Learning
3. Reinforcement Learning

Advantages of Machine learning:

1. Easily identifies trends and patterns.
2. no human intervention needed
3. continuous improvement
4. wide applications

Disadvantages of Machine Learning:

1. Time and resources
2. Data Acquisation


## Variables in Machine learning:-

Based on role in the model :--

1. Independent Variable (Feature / Input) :

- Variables used to make predictions.
- Denoted as X.
- Example: age, salary, pixels of an image.

2. Dependent Variable (Target / Output) :

- Variable the model predicts.
- Denoted as y.
- Example: house price, spam/not spam, disease yes/no.

Based on data type :--

1. Numerical (Quantitative) : Numbers with mathematical meaning.

- Continuous values: Can take infinite values.

  Example: height, temperature, price.

- Discrete Values:Can take Countable values.

  Example: number of students, clicks, items sold.

Based on categorical (quantitative) :--
* Represent categories or labels.

1. Nominal: No natural order.

    Example: color (red, blue), gender, country.

2. Ordinal: Ordered categories.

    Example: low < medium < high, ratings (1â€“5).

Based on label availability (learning type) :--

1. Labeled Variables : Input + output both available,
Used in supervised learning.

2. Unlabeled Variables : Only input data,Used in unsupervised learning.

Special variable types in ML practice :--

1. Binary Variables: Only two values.
    
    Example: 0/1, yes/no, true/false.

2. Boolean Variables: True or False.

3. Text Variables: Sentences, documents.
    Example: reviews, tweets.
    (need vectorization like TF-IDF, embeddings).

Image / Audio / Video Variables :--
* High-dimensional data.    
    
    Example: pixels, sound waves, frames.

Feature nature (advanced but important) :--

1. Raw Features: Directly collected data.

2. Engineered Features: Created from raw data.
    
    Example: BMI from height & weight.

## Data Cleaning

Data cleaning is the process of fixing or removing incorrect, incomplete, duplicate, or irrelevant data to improve model accuracy.

Q. Why data cleaning is important?

- Improves model performance.

- Reduces bias and noise.

- Prevents wrong predictions.

- Makes data ML-ready.

Steps to be followed in data cleaning :--

1. Data Inspection :

* Undestand data structurer.

* Check columns, data types, size,Identify errors.

2. Handle Missing Values  :

* Remove rows/columns.

* Fill with mean, median, mode.

* Forward / backward fill.

3. Remove Duplicate Records :

* Identify and delete repeated rows.

4. Correct Data Types :

* Convert incorrect formats.

* Example: string â†’ numeric, date â†’ datetime.

5. Handle Outliers :

* Detect extreme values.

* Remove, cap, or transform them.

6. Fix Inconsistent Data :

* Standardize text.

* Correct spelling and format.

* Example: "USA", "us", "United States".

7. Handle Noise :

* Smooth data.

* Remove irrelevant or random values.

8. Data Validation :

* Check logical rules.

* Example: age â‰¥ 0, salary > 0.

9. Clean Categorical Data :

* Merge rare categories.

* Correct labels.

10. Final Quality Check :

* Recheck missing values.

* Verify accuracy and consistency.

Note : These steps has to be choosen as per the dataset.Hence,not all steps are reqired to be followed Every time.

## Way to handle Missing Values

| Situation          | Best Method      |
| ------------------ | ---------------- |
| Small missing %    | Drop rows        |
| Numerical (skewed) | Median           |
| Numerical (normal) | Mean             |
| Categorical        | Mode / "Unknown" |
| Time-series        | Forward fill     |
| ML models          | KNN / Iterative  |
| Tree-based models  | No imputation    |


### Method 1: Delete the entire column.

```python
dataset = pd.read_csv('Dataset1.zip')
dataset.head(4)
```

```python
dataset.drop(columns=["Age"],inplace=True)
```

```python
dataset.shape
```

### Method 2: Removes all rows that contain at least one missing (NaN) values.

```python
dataset.dropna(inplace=True)
```

### Method 3: Fill the Missing values with a Value

```python
dataset.fillna(10,inplace=True)
```

```python
dataset.info()
```

### Method 4: Forward Filling (ffill)

Fills missing value with the previous valid value.

Example:-

Before:

* Day -  Sales
* 1   -  100
* 2   -  NaN
* 3   -  NaN
* 4   -  130


After forward fill:

* Day - Sales
* 1   - 100
* 2   - 100
* 3   - 100
* 4   - 130

### Code:

dataset.ffill(inplace=True)


### Method 5: Backward Filling (bfill)

Fills missing value with the next valid value.

Example:-

Before:

* Day -  Sales
* 1   -  100
* 2   -  NaN
* 3   -  NaN
* 4   -  130


After backward fill:

* Day  - Sales
* 1    - 100
* 2    - 130
* 3    - 130
* 4    - 130

### Code

dataset.bfill(inplace=True)

### Method 6: Mean,Median,Mode For Handling Missing values.

### Mean Imputation:

Replace missing values with the average.

Formula:

Mean=âˆ‘ð‘¥/ð‘›	â€‹

Example:-

Data: 10, 12, NaN, 14

Mean = (10 + 12 + 14) / 3 = 12

After imputation:

10, 12, 12, 14

### Code:

dataset['age'].fillna(dataset['age'].mean(), inplace=True)

### Median Imputation:

Replace missing values with the middle value.

Example:-

Data: 10, 12, NaN, 1000

Median = 12

After imputation:

10, 12, 12, 1000

### Code:

dataset['salary'].fillna(dataset['salary'].median(), inplace =True)

### Mode Imputation:

Replace missing values with the most frequent value.

Example:-

(categorical)

Red, Blue, Blue, NaN, Green


Mode = Blue

After imputation:

Red, Blue, Blue, Blue, Green

### Code:
dataset['city'].fillna(dataset['city'].mode()[0], inplace = True)
